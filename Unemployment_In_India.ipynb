### Unemployment In India

```python
#import Libarires
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
#read data
data=pd.read_csv("Unemployment_Rate_upto_11_2020.csv")
```

```python
#rename all the columns
data.columns=["States","Date","Frequency","Estimeted Unemployment Rate","Estimated Employed","Estimated Labour Participation Rate","Area","longitude","latitude"]
```

```python
 print(data)
```

```python
#look ata the correlation between the features of the dataset.
plt.figure(figsize=(12,10))
sns.heatmap(data.corr(),annot=True,cmap="YlGnBu")
plt.show()
```

```python
#look ata the estimated number of employes according to different region of india
data.columns=["States","Date","Frequency","Estimeted Unemployment Rate","Estimated Employed","Estimated Labour Participation Rate","Area","Latitude","longitude"]
plt.title("Indian Unemployment")
sns.histplot(x="Estimated Employed",hue="Area",data=data)
plt.show()
```

```python
#see the unemployment rate according to different regions of india

plt.figure(figsize=(12,10))
plt.title("Indian Umemployment")
sns.histplot(x="Estimeted Unemployment Rate",hue="Area",data=data)
plt.show()
```

```python
#Create a dashboard to analyze the unemployment rate of each indian state by region

unemploment = data[["States","Area","Estimeted Unemployment Rate"]]
figure=px.sunburst(unemploment,path=["Area","States"],values="Estimeted Unemployment Rate",
                   width=700,height=700,color_continuous_midpoint="RdYlGn",
                   title="Unemployment Rate in India")
figure.show()
```

```python
unemploment = data[["States","Area","Estimeted Unemployment Rate"]]
figure=px.bar(unemploment,["Area","States"],color="Estimeted Unemployment Rate")
figure.update_layout(xaxis={'categoryorder':'total descending'})
figure.show()
```
